```python
#Importuj bibliotekę numpy jako np oraz matplotlib.pyplot jako plt
import numpy as np
import matplotlib.pyplot as plt

# Generujmy losowe dane do użycia dla kul
data = np.random.randn(50, 2)

# Zdefiniujmy liczbę klastrów (k)
k = 4

# Zainicjujmy centra klastrów losowo
cluster_centers = np.random.randn(k, 2)

# Zdefiniujmy zmienną do śledzenia zmian w centrum
centers_changed = True

# Ustaw maksymalną liczbę iteracji, aby zapobiec pętli nieskończonej
max_iterations = 100

# Zdefiniujmy listę, aby przechowywać kolory dla każdego klastra
colors = ["green", "red", "yellow", "blue"]

# Zdefiniujmy zmienną, aby śledzić liczbę iteracji
iteration = 0

# Iterujmy, aż centra klastrów przestaną się zmieniać lub osiągnięto maksymalną liczbę iteracji
while centers_changed and iteration < max_iterations:
    # Zainicjujmy przydziały klastrów
    cluster_assignments = np.zeros(len(data))

    # Iterujmy przez każdy punkt danych - FIT
    for i, point in enumerate(data):
        # Zainicjuj najbliższą odległość od centrum klastra jako inf
        nearest_distance = np.inf

        # Iterujmy przez każde centrum klastra
        for j, center in enumerate(cluster_centers):
             # Oblicz odległość między punktem a centrum klastra
            distance = np.sqrt((point[0] - center[0])**2 + (point[1] - center[1])**2)
            
            #Jeśli ta odległość jest mniejsza niż aktualna najbliższa odległość, zaktualizuj najbliższą odległość i przydział klastra
            if distance < nearest_distance:
                nearest_distance = distance
                cluster_assignments[i] = j

    # Zapisz poprzednie centra klastrów
    prev_cluster_centers = np.copy(cluster_centers)

    # Iterujmy przez każdy klastr - PREDICT
    for i in range(k):
        # Znajdź wszystkie punkty danych przydzielone do tego klastra
        points = data[cluster_assignments == i]

        # Jeśli nie ma punktów przydzielonych do tego klastra, pomiń go
        if len(points) == 0:
            continue

        # Oblicz nowe centrum klastra, biorąc średnią ze wszystkich przydzielonych do niego punktów
        cluster_centers[i] = np.mean(points, axis=0)

    #Jeśli centra klastrów się nie zmieniły, zatrzymaj iterowanie
    if np.array_equal(prev_cluster_centers, cluster_centers):
        centers_changed = False

    # Zwiększ liczbę iteracji
    iteration += 1

# Narysuj punkty danych, kolorowane zgodnie z ich przydziałami do klastrów
for i in range(k):
    points = data[cluster_assignments == i]
    plt.scatter(points[:, 0], points[:, 1], c=colors[i])
    # Narysuj centra klastrów
    plt.scatter(cluster_centers[i, 0], cluster_centers[i, 1], c=colors[i], s=300, edgecolor="black", alpha=0.5)


plt.show()
```

```python
#Zaimportuj biblioteki
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

#Liczba centrów/klastrów
k = 4

#Wygenerujemy kilka losowych blobów za pomocą funkcji make_blobs:
X, y = make_blobs(n_samples=200, centers=k,random_state=8)

#Następnie możemy wyświetlić wygenerowane bloby za pomocą matplotlib:
#plt.scatter(X[:,0], X[:,1])

#Inicjalizacja modelu K-Means z k=4 klastrami
kmeans = KMeans(n_clusters=k)
#Dopasowanie modelu do danych oraz ustawienie ("Przewidzenie") klastrów dla każdego z punktów
pred_y = kmeans.fit_predict(X)


#Ustawienie odpowiednich kolorów dla klastrów
colors = ["green", "red", "yellow", "blue"]

#Na koniec możemy wyświetlić dane, tym razem kolorując punkty na podstawie ich przewidywanego skupiska:
for i in range(k):
  #Rysowanie punktów
  plt.scatter(X[pred_y==i,0], X[pred_y==i,1], color = colors[i])
  #Rysowanie centrów
  plt.scatter(kmeans.cluster_centers_[i][0], kmeans.cluster_centers_[i][1], color = colors[i], s=300, edgecolor="black", alpha=0.5)

#plt.scatter(X[:,0], X[:,1], c=pred_y)

#plt.scatter(X[pred_y==0,0], X[pred_y==0,1], color = "green")
#plt.scatter(X[pred_y==1,0], X[pred_y==1,1], color = "red")
#plt.scatter(X[pred_y==2,0], X[pred_y==2,1], color = "yellow")
#plt.scatter(X[pred_y==3,0], X[pred_y==3,1], color = "blue")

#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', edgecolor="black", alpha=0.5)
#plt.scatter(kmeans.cluster_centers_[0][0], kmeans.cluster_centers_[0][1], color = "green", s=300, edgecolor="black", alpha=0.5)
#plt.scatter(kmeans.cluster_centers_[1][0], kmeans.cluster_centers_[1][1], color = "red", s=300, edgecolor="black", alpha=0.5)
#plt.scatter(kmeans.cluster_centers_[2][0], kmeans.cluster_centers_[2][1], color = "yellow", s=300, edgecolor="black", alpha=0.5)
#plt.scatter(kmeans.cluster_centers_[3][0], kmeans.cluster_centers_[3][1], color = "blue", s=300, edgecolor="black", alpha=0.5)

plt.show()
```
